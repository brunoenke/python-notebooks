```python
#Exercícios em sala - Dictionary
#1-Carregue um arquivo texto e conte as palavras do arquivo, usando dictionary. 

#caminho("C:\\Users\\Bruno\\Documents\\Projects\\A alma do lazaro.txt")
file = open ("A alma do lazaro.txt")
palavras= dict()#vazio

import string
#para cada linha
for linha in file:
#para cada palavra na linha.split()
    linha = linha.translate(str.maketrans('','', string.punctuation))
    lista_palavras = linha.split()
    for palavra in lista_palavras:
        palavras[palavra] = palavras.get(palavra,0)+1
#print(palavras)
```

```python
#2-Imprima a palavra que mais aparece no arquivo (aquela que tem a contagem maior).
maior_valor = 0
maior_palavra = ''
for palavra, valor in palavras.items():
    if valor > maior_valor:
        maior_valor = valor
        maior_palavra = palavra
print ('A palavra que mais aparece no texto é {} {} vezes'.format(maior_palavra, maior_valor)) 
```

```python
#3-Gere um gráfico mostrando o número de vezes que cada palavra apareceu no arquivo (com matplotlib)
import matplotlib.pyplot as plt

eliminar = []
for key, value in palavras.items():
    if value < 100:
        eliminar.append(key)
    
for item in eliminar:
    del palavras[item]
        
plt.bar(palavras.keys(), palavras.values())
plt.show()
```

```python
#4-Carregue um arquivo texto e conte as letras que aparecem no arquivo. Considere:
#Converter tudo para lowercase
#Eliminar espaços, números, pontuação
#Teste com vários arquivos diferentes – veja como os resultados variam,
#compare os resultados de um arquivo com o de outro arquivo.
#As letras que mais se repetem são as mesmas?

frase = 'Contar quantas vezes a letra aparece no texto'
import string
file = open ("A alma do lazaro.txt")

letras= dict()
for linha in file:
    linha = linha.lower()
    linha = linha.translate(str.maketrans('','', string.punctuation))
    linha = linha.strip()
    for letra in linha:
        letras[letra] = letras.get(letra,0)+1
del letras[' ']        

maior_contador = 0
maior_letra = ''

for letra, valor in letras.items():
    if (valor> maior_contador):
        maior_contador = valor
        maior_letra = letra
print ('A letra que mais aparece no texto é {} {} vezes'.format(maior_letra, maior_contador)) 
```

```python
#5-Gere um gráfico de pizza usando as letras e sua respectiva contagem (com matplotlib)
import matplotlib.pyplot as plt

plt.bar(letras.keys(), letras.values())
plt.title(frase)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Blue', 'Orange', 'Green', 'Red'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Orange')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
