```python
#Quantas vezes a letra “M” aparece no arquivo “lorem.txt”?

# abrir o arquivo 
arquivo = open("lorem.txt")
# ler o arquivo
conteudo = arquivo.read()
# contar quantas vezes a letra M apareceu
quantidade = conteudo.upper().count("M")
print("A letra M aparece {} vezes no arquivo".format(quantidade))
```

```python
#Quantas vezes a palavra “lorem” aparece no mesmo arquivo?
# abrir o arquivo 
arquivo = open("lorem.txt",'r')
# ler o arquivo
conteudo = arquivo.read()
# contar quantas vezes lorem apareceu
quantidade = conteudo.lower().count("lorem")

print("A palavra “lorem” aparece {} vezes no arquivo".format(quantidade))
```

```python
#Quais letras do alfabeto não aparecem no arquivo “lorem.txt”?
arquivo = open("lorem.txt", 'r')
# ler o arquivo
conteudo = arquivo.read()

#importar contantes do alfabeto
from string import ascii_uppercase, ascii_lowercase, ascii_letters

for char in ascii_letters:
    #contar
    contador = conteudo.count(char)
    if (contador == 0):
        print('A letra {} não está no arquivo lorem.txt'.format(char))
```

```python
#Qual letra maiúscula aparece mais vezes no arquivo “lorem.txt”?

#Abrir o arquivo
arquivo = open("lorem.txt", 'r')
#Ler o arquivo
conteudo = arquivo.read()

maior_contador = 0
maior_letra = ''

#importar contantes do alfabeto
from string import ascii_uppercase, ascii_lowercase, ascii_letters

for letra in ascii_uppercase:
    #contar quantas vezes a letra atual aparece no arquivo
    contador = conteudo.count(letra)
    
    #saber qual a letra que mais aparece (maior boi)
    if (contador> maior_contador):
        maior_contador = contador
        maior_letra = letra
print ('A letra {} é quem aparece mais vezes no arquivo'.format(maior_letra, maior_contador)) 
    
```

```python
#Gere um arquivo com a tabuada de 10 a 20, com as colunas alinhadas à esquerda e 
#separadores entre cada grupo de 10 com título centralizado.

tabuada = open("tabuada.txt","w")
count = 10
#para cada tabuada
while count <= 20:
    titulo = "Tabuada de {}".format(count) 
        
    tabuada.writelines("{:*^50}".format(titulo))
    tabuada.writelines("\n")
    
    fator = 1
    while fator <= 10:
        linha = "{} * {} = {}".format(count, fator, count*fator)
        tabuada.writelines("{:<50}".format(linha))
        tabuada.writelines("\n")
        fator += 1
    #proxima tabuada
    count += 1

tabuada.close()
```

```python
#Crie um arquivo contendo os números de 100 a 1000000 em formato hexadecimal.

file = open('numerosHex.txt', 'w')
contador = 100
while contador <= 1000000:
    conteudo = ('Dec {0:d}; Hex: {0:x} \n'.format(contador))
    contador += 1
    file.write(conteudo)
file.close()
```

```python
#Leia o arquivo criado anteriormente e grave um novo arquivo com o mesmo conteúdo, 
#eliminando todos os números inteiros terminados em zero.

#Ler arquivo
file = open('numerosHex.txt','r')
#Gravar novo arquivo
new_file = open ('novo_arquivo.txt','w')
#Modificar conteudo, eliminando linhas cujo número
for linha in file:
    #pegar o primeiro campo da linha que é decimal
    numero_decimal = linha [0:7]
    #tirar os espaços da string strip()
    numero_decimal = numero_decimal.strip()
    #verificar se termina com zero
    if not (numero_decimal.endswith('0')):
        new_file.writelines(linha)
file.close()
new_file.close()
```
