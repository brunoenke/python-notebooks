```python
#Carregue um arquivo texto
#conte as palavras do arquivo usando dictionary

#caminho("C:\\Users\\Bruno\\Documents\\Projects\\A alma do lazaro.txt")
file = open ("A alma do lazaro.txt")
palavras= dict()#vazio

import string
#para cada linha
for linha in file:
#para cada palavra na linha.split()
    linha = linha.translate(str.maketrans('','', string.punctuation))
    lista_palavras = linha.split()
    for palavra in lista_palavras:
        palavras[palavra] = palavras.get(palavra,0)+1
```

```python
#imprima as 10 palavras que mais aparecem no arquivo (aquelas que tem a contagem maior) usando tuplas.
import matplotlib.pyplot as plt

lista = []
for palavra, contador in palavras.items():
    lista.append((contador,palavra))

lista.sort(reverse=True)
result = lista[:10]
keylist = []
valuelist = []

for contador,palavra in result:
    keylist.append(palavra)
    valuelist.append(contador)

plt.bar(keylist, valuelist)
plt.show()
```

```python
#Escreva um programa que leia um arquivo e imprima as letras em ordem
#decrescente de frequência as letras com maior ocorrência primeiro.
import string
import matplotlib.pyplot as plt

def carregarArquivo(caminho):
    file = open(caminho, encoding="ANSI")
    letras = dict() #vazio
    #para cada linha no arquivo 
    for linha in file:
        linha = linha.lower().translate(str.maketrans('', '', string.punctuation+' '+string.digits))
        #para cada letra na string
        for letra in linha:
            letras[letra] = letras.get(letra,0) + 1
    return letras

def imprimirGrafico(caminho):
    letras = carregarArquivo(caminho)
    lista = []
    for letra, contador in letras.items():
        lista.append((contador,letra))
    lista.sort(reverse=True)
    keylist = []
    valuelist = []
    for contador,letra in lista:
        keylist.append(letra)
        valuelist.append(contador)
    plt.bar(keylist, valuelist)
    return plt.show()
  
imprimirGrafico("A alma do lazaro.txt")
imprimirGrafico("O banqueiro anarquista.txt")
imprimirGrafico("Papeis Avulsos.txt")
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(valuelist, labels=keylist, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
